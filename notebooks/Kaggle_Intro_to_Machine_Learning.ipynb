```python
import pandas as pd
# Display all columns when printing the DataFrame
pd.set_option('display.max_columns', None)
```

```python
melbourne_data = pd.read_csv('/content/melb_data.csv')
melbourne_data
```

```python
melbourne_data.describe()
```

```python
melbourne_data.info()
```

```python
#Missing values
print(melbourne_data.isna().sum())

#Clean all missing values
melbourne_data = melbourne_data.dropna(axis=0)

# Remove duplicated rows from the entire DataFrame
melbourne_data_clean = melbourne_data.drop_duplicates()

```

```python
#Selecting Data for Modeling
#Selecting The Prediction Target
y = melbourne_data.Price
```

```python
#Choosing "Features". We'll build a model with only a few features
melbourne_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
X = melbourne_data[melbourne_features]
X.describe()
```

```python
X.head()
```

```python
#Building Model
from sklearn.tree import DecisionTreeRegressor

# Define model. Specify a number for random_state to ensure same results each run
melbourne_model = DecisionTreeRegressor(random_state=1)

# Fit model
melbourne_model.fit(X, y)
```

```python
print("Making predictions for the following 5 houses:")
print(X.head())
print("The predictions are")
print(melbourne_model.predict(X.head()))
```

```python
#Model Validation
from sklearn.metrics import mean_absolute_error
#Mean Absolute Error MAE
predicted_home_prices = melbourne_model.predict(X)
mae = mean_absolute_error(y, predicted_home_prices)
print(mae)
print(f'On average, the model makes an error of approximately ${mae:.2f} when predicting the price of a house.')
```

```python
#Validation data
#We'll use some of that data as training data to fit the model, and we'll use the other data as validation data to calculate mean_absolute_error.
from sklearn.model_selection import train_test_split

# split data into training and validation data, for both features and target
# The split is based on a random number generator. Supplying a numeric value to
# the random_state argument guarantees we get the same split every time we
# run this script.
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 1)
# Define model
melbourne_model = DecisionTreeRegressor(random_state = 1)
# Fit model
melbourne_model.fit(train_X, train_y,)

# get predicted prices on validation data
val_predictions = melbourne_model.predict(val_X)
print(mean_absolute_error(val_y, val_predictions))
new_mae = mean_absolute_error(val_y, val_predictions)
print(f'On average, the model makes an error of approximately ${new_mae:.2f} when predicting the price of a house.')
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor

#We train the decision tree model with a fixed number of leaf nodes and calculate the MAE.
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return(mae)
```

```python
#Compare MAE with differing values of max_leaf_nodes
for max_leaf_nodes in [5, 50, 500, 5000]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
#For this case, the best option is 500 nodes (MAE: 243613)
```

##Randon Forest

```python
# Choose target and features
y = melbourne_data.Price
melbourne_features = ['Rooms', 'Bathroom', 'Landsize', 'BuildingArea',
                        'YearBuilt', 'Lattitude', 'Longtitude']
X = melbourne_data[melbourne_features]

from sklearn.model_selection import train_test_split

# split data into training and validation data, for both features and target
# The split is based on a random number generator. Supplying a numeric value to
# the random_state argument guarantees we get the same split every time we
# run this script.
train_X, val_X, train_y, val_y = train_test_split(X, y,random_state = 0)
```

```python
# We build a random forest model similarly to how we built a decision tree in scikit-learn
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

forest_model = RandomForestRegressor(random_state=1)
forest_model.fit(train_X, train_y)
melb_preds = forest_model.predict(val_X)
print(mean_absolute_error(val_y, melb_preds))
```
